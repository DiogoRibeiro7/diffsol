# diffsol-pytorch tutorial

This notebook demonstrates the high-level API. The heavy lifting is performed by the Rust bindings.

```python
import torch
import diffsol_pytorch as dsp

code = """
state x
param k
der(x) = -k * x
"""

module = dsp.DiffsolModule(code)
params = torch.tensor([0.1], dtype=torch.float64)
times = torch.linspace(0, 1, 5, dtype=torch.float64)
module.solve_dense(params.tolist(), times.tolist())
```
